```
import tarfile
import os
import urllib.request

def downloadDataset():
  file = urllib.request.urlretrieve('http://download.tensorflow.org/example_images/flower_photos.tgz', 'flower_photos.tgz')
  tar = tarfile.open('flower_photos.tgz')
  tar.extractall("")
  tar.close()
```

```
def createFolder():
  current_directory = os.getcwd()
  final_directory = os.path.join(current_directory, r'tf_files')
  if not os.path.exists(final_directory):
   os.makedirs(final_directory);
   print("folder has been created");
  else:
    print("folder is already exist");
```

```
createFolder();
downloadDataset();
```
